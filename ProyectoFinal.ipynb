### **Gráficas**

```python
plt.figure(figsize=(20, 10))
frecuencia_paises = df['País de Residencia'].value_counts().nlargest(20)
ax = frecuencia_paises.plot(kind='bar', color='skyblue', edgecolor='black')

umbral = 50
for p in ax.patches:
    if p.get_height() > umbral:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points', rotation=45)

plt.xticks(rotation=45, ha='right')
plt.title('Frecuencia de País')
plt.xlabel('País')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

### **País de Residencia de los visitantes más frecuentes**
  En el gráfico encontramos los 20 paises de origen más frecuentes de los visitantes no residentes que ingresan al país por los diferentes puntos migratorios,excluyendo a a los extranjeros con residencia venezolana debido al efecto migratorio.
  
  Los primeros paises en el grafico son Estados Unidos, Ecuador y México, tienen una frecuencia mayor en comparación con los demás paises.
  
  Hay una disminución gradual en la frecuencia a medida que se avanza hacia la derecha en el gráfico, lo que indica que hay paises con muy pocos registros.
  
  Estados Unidos es el País de Residencia de los visitantes más frecuentes.

```python
plt.figure(figsize=(20, 10))
frecuencia_departdes = df['Departamento'].value_counts()
ax = frecuencia_departdes.plot(kind='bar', color='skyblue', edgecolor='black')

umbral = 50
for p in ax.patches:
    if p.get_height() > umbral:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points', rotation=45)

plt.xticks(rotation=45, ha='right')
plt.title('Departamento de Destino ')
plt.xlabel('Departamento')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

### **Departamentos más frecuentados por visitantes no recidentes**

El departamento de destino más visitado es: Antioquia, seguido de el Valle del Cauca y Cundinamarca.

Hacia la derecha del gráfico vemos como disminuye, indicando que los departamentos de Vapués, Guaínia y Guaviare son los menos visitados.

```python
plt.figure(figsize=(20, 10))
frecuencia_año = df['Año'].value_counts()
ax = frecuencia_año.plot(kind='bar', color='skyblue', edgecolor='black')

umbral = 50
for p in ax.patches:
    if p.get_height() > umbral:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points', rotation=45)

plt.xticks(rotation=45, ha='right')
plt.title('Año más frecuentado')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

### **Año con más visitas**
En el gráfico podemos evidenciar que el año con más visitas desde el 2015 al mes de agosto de 2024 fue el 2023.

El año con menos frecuencia de visitantes fue el 2020, podemos deducir que este resultado se debe a la crisis mundial que se vivió ese año, donde no era posible viajar.

También vemos la recuperación que tuvo el país en temas de turismo después de la pandemia.

```python
frecuencia_mes = df['Mes'].value_counts()
plt.figure(figsize=(20, 10))
plt.pie(frecuencia_mes, labels=frecuencia_mes.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribución de la Frecuencia por Mes')
plt.axis('equal')
plt.show()
```

### **Mes con más frecuencia de visitantes**
El gráfico no muestra que el mes número 12 (diciembre) es el mes en el cual el país recibe más visitas por ciudadanos no residentes, seguido del mes 1 (enero) y mes 2 ( febrero). Podemos deducir que esta epoca es usualmente temporada alta debido a las vacaciones lo que incrementa el número de visitantes.

Al contrario, el mes con menos visitas es el 5 (mayo) un mes de temporada baja, en el cual normalmente es laboral.


```python
mejor_mes_por_año = df.groupby(['Año', 'Mes'])['Extranjeros no Residentes'].sum().reset_index()
mejor_mes_por_año = mejor_mes_por_año.loc[mejor_mes_por_año.groupby('Año')['Extranjeros no Residentes'].idxmax()]
print(mejor_mes_por_año)

plt.figure(figsize=(14, 8))
plt.plot(mejor_mes_por_año['Año'].astype(str) + '-' + mejor_mes_por_año['Mes'].astype(str),
         mejor_mes_por_año['Extranjeros no Residentes'], marker='o', linestyle='-', color='b')
plt.title('Mejor mes de cada año')
plt.xlabel('Año-Mes')
plt.ylabel('Extranjeros no Residentes')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### **El mejor mes de cada año**

En la gráfica evidenciamos que el mes con mayor número de visitantes extranjeros fue febrero del 2024. Tambien podemos deducir que el mes 12 (diciembre) es el mejor mes de la mayoria de los años.
